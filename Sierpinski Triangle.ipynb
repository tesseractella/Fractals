The Sierpinski Triangle is defined by recursively splitting an equilateral tringle into 4  smaller equlateral triangles and removing the middle one. It is an attractor 

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def Sierpinski(ax, depth, x, y, size):
    if depth == 0:
        points = [[x, y], [x+size, y], [x+size/2, y+size]]
        ax.fill([p[0] for p in points], [p[1] for p in points], 'b')
    else:
        Sierpinski(ax, depth-1, x, y, size/2)
        Sierpinski(ax, depth-1, x+size/2, y, size/2)
        Sierpinski(ax, depth-1, x+size/4, y+size/2, size/2)
```

```python
depth = 5
fig, ax = plt.subplots()
Sierpinski(ax, depth, 0, 0, 32)
plt.show()
```
